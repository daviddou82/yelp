```python
import json
import os
import pandas as pd
import ast
import time
from textblob import TextBlob
import re
import spacy
nlp = spacy.load('en_core_web_lg')
pd.options.display.max_columns = None
```

```python
df_user=pd.read_csv("user.csv")
```

```python
df_review=pd.read_csv("review.csv")
```

```python
df_business=pd.read_csv("business.csv")
```

```python
df_review =df_review.merge(df_business,on="business_id")
```

```python
review_chit = df_review.loc[df_review['name']=="Chipotle Mexican Grill"]
```

```python
string = ""
for index ,row in review_chit.iterrows():
    review_text = review_chit.at[index,'text']
    string += review_text
string = string[:999999]
```

```python
# print (string[:1300])
pos_counts = defaultdict(Counter)
doc = nlp(string)

for token in doc:
    pos_counts[token.pos][token.orth] += 1

for pos_id, counts in sorted(pos_counts.items()):
    pos = doc.vocab.strings[pos_id]
    for orth_id, count in counts.most_common():
        print(pos, count, doc.vocab.strings[orth_id])
```
