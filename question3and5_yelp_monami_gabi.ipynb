```python
import json
import os
import pandas as pd
import ast
import time
from textblob import TextBlob
import re
import spacy
nlp = spacy.load('en_core_web_lg')
pd.options.display.max_columns = None
```

```python
df_user=pd.read_csv("./data/user.csv")
```

```python
df_review=pd.read_csv("./data/review.csv")
```

```python
df_business=pd.read_csv("./data/business.csv")
```

```python
df_review.head(1)
```

```python
# join 
df_review =df_review.merge(df_business,on="business_id")
```

```python
review_last_year =df_review[(df_review['date'] > '2017-1-1 01:00:00') & (df_review['date'] <= '2018-1-1 04:00:00')]
```

```python
review_canada = review_last_year.loc[review_last_year.postal_code.str.len()==7].reset_index()
review_canada_agg = review_canada.groupby(['user_id']).count().reset_index()
review_canada_agg = review_canada_agg.sort_values(by=['index'],ascending=False)
review_canada_agg = review_canada_agg[review_canada_agg['index']>=10]
```

```python
review_mon = review_last_year.loc[review_last_year['name']=="Mon Ami Gabi"] 
```

```python
review_mon_canadian =review_mon.merge(review_canada_agg,on="user_id")
```

```python
# Find  users from Ontario who have left more than 10 reviews.
review_on = df_review.loc[df_review['state']=="ON"]
review_on_agg = review_on.groupby(['user_id']).count().reset_index()
review_on_agg = review_on_agg.sort_values(by=['state'],ascending=False)
review_on_agg = review_on_agg[review_on_agg['state']>=10]
```

```python
review_mon = df_review.loc[df_review['name']=="Mon Ami Gabi"]
```

```python
review_ontario10_mon =review_mon.merge(review_on_agg,on="user_id").head()
```
